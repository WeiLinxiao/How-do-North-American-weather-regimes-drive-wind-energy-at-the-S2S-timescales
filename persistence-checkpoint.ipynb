```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import xarray as xr
import numpy as np
from glob import glob
from yllib.function import mysom
from yllib.function import myfunc
from yllib.plot import plot_map
from cartopy import crs as ccrs
import matplotlib.pyplot as plt
import matplotlib as mpl
%config InlineBackend.figure_format='retina'
%matplotlib inline
```

```python
year1   = 1981
year2   = 2020
cold_months  = [10,11,12,1,2,3]
warm_months  = [4,5,6,7,8,9]
mapsize = (4,4)
factor  = ('z')
map_extent = [160,-20+360,25,90]
disp_order_cold = [1,0,3,2]
disp_order_warm = [3,1,2,0]
```

```python
bmu_cold = xr.open_dataset(f"som/kmean_cold_z_{mapsize[0]}x{mapsize[1]}.nc")['bmu']
bmu_warm = xr.open_dataset(f"som/kmean_warm_z_{mapsize[0]}x{mapsize[1]}.nc")['bmu']
```

```python
data = xr.open_dataset('data/hgt_daily_anomaly_globe.nc')
# select season
data_cold = data.sel(time=data['time.month'].isin(cold_months))['z500']
data_warm = data.sel(time=data['time.month'].isin(warm_months))['z500']
```

```python
def calc_persistence(bmu, data, season):
    bmu.coords["time"] = data.coords["time"]

    from itertools import groupby
    pers = []
    for ibmu in np.arange(4):
        a = bmu.where(bmu==ibmu)
        result = [list(v) for k,v in groupby(a,np.isfinite) if k]
        result = [xr.concat(v, dim="time") for v in result]
        count = [v.size for v in result]
        time  = [v[0].coords["time"].data for v in result]
        count = xr.DataArray(count, dims=["time"], coords=[time], name="pers")
        pers.append(count)
        count.to_netcdf(f'som/som_persistence_{season}_{ibmu}.nc')
    return pers
```

### Calculate persistence

```python
pers_cold = calc_persistence(bmu_cold, data_cold, 'cold')
pers_warm = calc_persistence(bmu_warm, data_warm, 'warm')
```

### Load persistence

```python
pers_cold = [pers_cold[i] for i in disp_order_cold]
pers_warm = [pers_warm[i] for i in disp_order_warm]
```

```python
def disp_statistics(pers):
    pers = [np.asarray([[p]*p for p in list(per.data)]).sum() for per in pers]
disp_statistics(pers_cold)
disp_statistics(pers_warm)
```

```python
def plot_persistence(ax, pers, season):
    pers = [np.asarray([[p]*p for p in list(per.data)]).sum() for per in pers]

    from sklearn.neighbors import KernelDensity
    bw = []
    for per in pers:
        per = np.array(per)
        miny, maxy = per.min(), per.max()
        kde = KernelDensity(kernel='gaussian', bandwidth=1).fit(per[:, np.newaxis])
        fit = np.exp(kde.score_samples(np.linspace(miny, maxy, 1000)[:, np.newaxis]))
        bw.append(np.max(fit))
    bw = np.array(bw)
    bw = bw/bw.max()

    vp = ax.violinplot(pers, np.array([0,1,2,3])-0.06, widths=1*bw,
               showmeans=False, showmedians=False, showextrema=False)

    for patch in vp['bodies']:
        patch.set_edgecolor('tab:blue')
        patch.set_alpha(0.4)

    # Clip the right half of each violin.
    for item in ax.collections:
        x0, y0, width, height = item.get_paths()[0].get_extents().bounds
        item.set_clip_path(plt.Rectangle((x0, y0), width/2, height,
                           transform=ax.transData))
    #        plt.setp(ax.collections, alpha=.3)

    box  = ax.boxplot(pers, positions=np.array([0,1,2,3])+0.06, widths=0.2, sym="x", patch_artist="true", showfliers=True)

    for patch in box["boxes"]:
        patch.set_facecolor("tab:gray")
        patch.set_edgecolor("white")
    for patch in box["medians"]:
        patch.set_color("white")
    for item in ["caps", "whiskers"]:
        for patch in box[item]:
             patch.set_color("tab:gray")
    for patch in box["fliers"]:
        patch.set_markeredgecolor("tab:gray")
        patch.set_markersize(3)

    types = ['AkR', 'PT', 'ArH', 'ArL']

    ax.set_ylim(-5, 90)
    ax.set_ylabel("Persistence (day)", fontsize=12)
    ax.set_xticks([0,1,2,3])
    ax.set_xticklabels(types, fontsize=12)
# plt.savefig(f"pdf/persistence_{season}.pdf", bbox_inches="tight", transparent=True)
# plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,4))
plt.subplots_adjust(left=0.15, bottom=0.15, right=0.98, top=0.95, hspace=0.2, wspace=0.2)
axs = axs.ravel()
plot_persistence(axs[0], pers_cold, 'cold')
plot_persistence(axs[1], pers_warm, 'warm')
axs[0].set_title('(a) Cold season', loc='left', fontsize=14)
axs[1].set_title('(b) Warm season', loc='left', fontsize=14)
```
